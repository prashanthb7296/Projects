```python
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
data = pd.read_csv("IRIS.csv")
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe().T
```

```python
data.isnull().any().sum()
```

```python
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
```

```python
data.shape
```

```python
data["species"].value_counts()
```

```python
#Multi-variate Analysis
data.plot(x='petal_length', y='petal_width', kind='scatter')
plt.show()
```

Multivariate Analysis

```python
sns.set_style("whitegrid")
sns.FacetGrid(data,height=4,hue="species").map(plt.scatter,"petal_length","petal_width").add_legend()
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(data,height=4,hue="species").map(plt.scatter,"sepal_length","sepal_width").add_legend()
plt.show()
```

Bivariate Analysis

```python
sns.pairplot(data,height=3,hue='species')
plt.show()
```

```python
sns.FacetGrid(data,hue="species",height=4).map(sns.distplot,"petal_length").add_legend()
plt.show()
```

```python
sns.FacetGrid(data,hue="species",height=4).map(sns.distplot,"petal_width").add_legend()
plt.show()
```

```python
sns.FacetGrid(data,hue="species",height=4).map(sns.distplot,"sepal_width").add_legend()
plt.show()
```

```python
sns.FacetGrid(data,hue="species",height=4).map(sns.distplot,"sepal_width").add_legend()
plt.show()
```

Outlier and Data Distribution

```python
sns.boxplot(data=data,x='species',y='petal_length')
plt.show()
```

```python
sns.violinplot(data=data,x="species",y="petal_length")
plt.show()
```

```python
sns.jointplot(data=data,x="petal_length",y="petal_width",kind="reg",height=4)
plt.show()
```

```python
#for categorical variable
sns.countplot(x="species",data=data)
plt.show()
```

```python
#for numerical variable
sns.histplot(data=data,x='sepal_width')
plt.show()
```

Correlation Data Analysis

```python
corr = data.corr()
corr
```

```python
sns.heatmap(data=corr,annot=True,vmin=-1,vmax=1,center=0)
plt.show()
```

```python
#categorical variables
labels = ['Iris-versicolor','Iris-virginica','Iris-setosa']
explode = (0.1, 0.0, 0.1)
colors = ("cyan","indigo","orange")
plt.pie(x=data["species"].value_counts(),labels=labels,explode=explode,colors=colors)
plt.legend(labels,loc='best')
plt.show()
```
